```python
import numpy as np
import pandas as pd

# Importing the dataset
data = pd.read_csv('H:\\MCA\\3rd Sem\\Python\\example_data.csv')
data.head(10)
```

```python
# Cleaning the data
# For example, removing missing values:
data = data.dropna()
data.head()
```

```python
# Data frame manipulation using NumPy
# For example, scaling the values of a column to be between 0 and 1:
column_to_scale = data['age']
scaled_column = (column_to_scale - np.min(column_to_scale)) / (np.max(column_to_scale) - np.min(column_to_scale))
data['age'] = scaled_column

# Printing the updated data frame
print(data.head())
```
