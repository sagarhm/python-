```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate some time-series data
dates = pd.date_range('20210101', periods=365)
ts = pd.Series(np.random.randn(len(dates)), index=dates)

# Print the time series data
print("Time series data:")
print(ts)
```

```python
# Plot the time series data
plt.plot(ts)
plt.title("Time Series Plot")
plt.xlabel("Date")
plt.ylabel("Value")
plt.show()
```

```python
# Resample the time series data to a monthly frequency
monthly_ts = ts.resample('M').mean()

# Print the resampled time series data
print("Monthly time series data:")
print(monthly_ts)
```

```python
# Plot the resampled time series data
plt.plot(monthly_ts)
plt.title("Monthly Time Series Plot")
plt.xlabel("Date")
plt.ylabel("Value")
plt.show()
```

```python
# Compute the rolling mean of the time series data
rolling_mean = ts.rolling(window=30).mean()

# Print the rolling mean data
print("Rolling mean data:")
print(rolling_mean)
```

```python
# Plot the rolling mean data
plt.plot(ts, label='Original Data')
plt.plot(rolling_mean, label='Rolling Mean')
plt.title("Rolling Mean Plot")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()
```
