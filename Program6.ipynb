```python
import numpy as np
import matplotlib.pyplot as plt

# Create sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
# Line plot
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Scatter plot
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Bar chart
labels = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 30, 40, 50]
plt.bar(labels, values)
plt.title("Bar Chart")
plt.xlabel("Labels")
plt.ylabel("Values")
plt.show()
```

```python
# Histogram
data = np.random.normal(0, 1, size=1000)
plt.hist(data, bins=30)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pie chart
labels = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 30, 40, 50]
plt.pie(values, labels=labels)
plt.title("Pie Chart")
plt.show()
```
