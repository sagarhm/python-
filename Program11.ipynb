```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Display the first 5 rows of the dataset
print("First 5 rows of the iris dataset:")
print(iris.head())
```

```python
# Create a scatterplot of the iris dataset
sns.scatterplot(x="sepal_length", y="petal_length", data=iris, hue="species")

# Add a title to the plot
plt.title("Scatterplot of Sepal Length vs. Petal Length")

# Show the plot
plt.show()
```

```python
# Create a histogram of the iris dataset
sns.histplot(data=iris, x="petal_width", hue="species")

# Add a title to the plot
plt.title("Histogram of Petal Width")

# Show the plot
plt.show()
```

```python
# Create a boxplot of the iris dataset
sns.boxplot(x="species", y="sepal_width", data=iris)

# Add a title to the plot
plt.title("Boxplot of Sepal Width by Species")

# Show the plot
plt.show()
```

```python
# Create a violin plot of the iris dataset
sns.violinplot(x="species", y="petal_length", data=iris)

# Add a title to the plot
plt.title("Violin Plot of Petal Length by Species")

# Show the plot
plt.show()
```
